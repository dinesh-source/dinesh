```python
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator

```

```python
train_datagen = ImageDataGenerator(
    rescale=1.0 / 255,       # Rescale pixel values to [0, 1]
    shear_range=0.2,         # Shear transformation
    zoom_range=0.2,          # Random zoom
    horizontal_flip=True    # Random horizontal flip
)

# Load and augment training data using the generator
train_generator = train_datagen.flow_from_directory('train',
    # train_data_dir,
    target_size=(64,64),  # Resize images to (224, 224)
    batch_size=32,
    class_mode='binary' # Assuming you have categorical labels
)

test_datagen = ImageDataGenerator(
    rescale=1.0 / 255,)

test_set=test_datagen.flow_from_directory('valid',
    # train_data_dir,
    target_size=(64,64),  # Resize images to (224, 224)
    batch_size=32,
    class_mode='binary' # Assuming you have categorical labels
)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
cnn=Sequential()
```

```python
cnn.add(Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[64, 64, 3]))
```

```python
cnn.add(Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tf.Flatten())
```

```python
cnn.add(tf.keras.layers.Dense(units=128, activation='relu'))
cnn.add(tf.keras.layers.Dense(units=128, activation='tanh'))
```

```python
cnn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history=cnn.fit(x = train_generator, validation_data = test_set,epochs = 3)

```
